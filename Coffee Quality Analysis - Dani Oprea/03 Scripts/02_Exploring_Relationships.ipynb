#### Update Grading Date column to datetime

```python
# Remove ordinal suffixes from the Grading Date column
df['Grading Date'] = df['Grading Date'].str.replace(r'st|nd|rd|th', '')

# Convert the grading Date column to datetime
df['Grading Date'] = df['Grading Date'].apply(lambda x: parser.parse(x))
```

#### Create a correlation heatmap

```python
df.columns
```

```python
# Create a subset for the heatmap

sub = df[['Altitude','Grading Date','Aroma', 'Flavor','Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Overall','Total Cup Points','Moisture Percentage', 'Category One Defects','Quakers', 'Category Two Defects']]
```

```python
sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(14,14))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
heatmap = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

To begin, we will ignore the correlations between aroma up through total cup points. It is safe to say that a good quality coffee will have high scores across all these categories, so that is why there exists such high correlation. The uniformity column only has two ratings outside of a solid 10, so we can ignore that column for the most part as it doesn't provide much variance. From here, there aren't any massive correlations, but there are still some worth looking into.

First, there does seem to be a medium positive correlation between quakers and category two defects (i.e. more quakers, more category two defects). We will look into this further.

Next, there does seem to be a weak-to-medium negative correlation between total cup points (as well as most sensory evaluations) and category two defects (more category two defects, the lower the total cup rating). The same weak-to-medium negative correlation exists between quakers and total cup rating. We will look into this further.

There seems to be a weak positive correlation between altitude and body, and possibly category two defects.

There are definitely some insights to be gleaned from these scatterplots.

First off, we can see pretty quickly there is not much meaningful correlation between altitude and quakers. There are many quakers around the 1200 altitude, moreso than at higher altitudes. Therefore, we can say a correlation does not exist.

We could say the same for body and altitude. There are higher body ratings around 1800-2000 altitude than at higher altitudes, so while a linear correlation does not exist, there may be a polynomial regression here. It could be that 1800-2000 altitude results in better quality coffee. This may be worth exploring more, possibly with a polynomial regression analysis.

Next, there does seem to be a week correlation between quakers and category two defects. There are more quakers at higher category two defects ratings however, the highest category two defect (16) has zero quakers. So I would say the correlation is weak at best, but does seem to exist.

Lastly, there does seem to be a weak negative correlation between quakers and category two defects vs total cup points. There is an anomaly where a quaker count of zero and a category two defect of one resulted in one of the lowest total cup ratings (78) however, outside of this, there does seem to be a trend that the presence of more quakers and category two defects results in lower scores.

#### Categorical Plots

```python
# Use a histogram to visualize the distribution of total cup points

sns.histplot(df['Total Cup Points'], bins = 20, kde = True)
```

```python
# Create total cup rating categories

df.loc[df['Total Cup Points'] < 82, 'Rating Category'] = 'Low Rating'
```

```python
df.loc[(df['Total Cup Points'] >= 82) & (df['Total Cup Points'] < 85.5), 'Rating Category'] = 'Middle Rating'
```

```python
df.loc[df['Total Cup Points'] >= 85.5, 'Rating Category'] = 'High Rating'
```

```python
df['Rating Category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the rating categories created above vs country of origin

sns.set(style="ticks")
g1 = sns.catplot(x='Country of Origin', y='Total Cup Points', hue="Rating Category", data=df, height=7, width=7)
plt.xticks(rotation=90)
g1.set_axis_labels("Country of Origin", "Total Cup Points") 
```

This categorical plot gives us quite a bit of detail about output the highest and lowest quality coffees. It is clear that Colombia produces the highest rated coffee. However, Taiwan produces the largest amount of highest quality coffees. Taiwan also seems to be a bit more consistent than Colombia when it comes to quality of coffee. There is quite a lot of spread between coffee ratings in Colombia, whereas in Taiwan, they're a bit more grouped. 

It looks like Brazil and El Salvador produce the lowest rated coffees of the group. 

```python
# Create a categorical plot in seaborn using the rating categories created above vs processing method

sns.set(style="ticks")
g1 = sns.catplot(x='Processing Method', y='Total Cup Points', hue="Rating Category", data=df, height=5, width=5)
plt.xticks(rotation=90)
g1.set_axis_labels("Processing Method", "Total Cup Points") 
```

This categorical plot offers some detail about processing method vs total rating but it seems like it should be taken with a grain of salt due to quantities present in the plot. It seems that the Double Anaerobic Washed method leads results in the highest quality coffee, however there is only one coffee in our analysis that uses this processing method. Therefore, we would need to look at more data on this processing method to truly be able to say whether or not it results in the best quality coffee.

From there, it looks like the Washed / Wet method leads to the second highest ratings. It's important to note that there is a lot of spread in this method. There are certainly still lower quality coffees using this method. Overall though. it seems that this method has the largest amount of high rated coffees compared to others. 

The Natural / Dry method looks to also produce highly rated coffee though also contains one of the lowest rated coffees as well.

```python
# Create a categorical plot in seaborn using the rating categories created above vs variety

sns.set(style="ticks")
g1 = sns.catplot(x='Variety', y='Total Cup Points', hue="Rating Category", data=df, height=8, width=10)
plt.xticks(rotation=90)
g1.set_axis_labels("Variety", "Total Cup Points") 
```

Lastly, this final categorical plot also provides some solid insight into variety vs rating. It seems that Castillo variety leads to the highest rated coffee. However, like the processing method plot, we only have one coffee of this variety in our analysis. Therefore, we would need to analyze more coffees of this variety type to see if that assumption is true. 

From here, it looks like Gesha variety produces the largest amount of highly rated coffees. 

```python
# Create a categorical plot in seaborn using the rating categories created above vs harvest year

sns.set(style="ticks")
g1 = sns.catplot(x='Harvest Year', y='Total Cup Points', hue="Rating Category", data=df)
plt.xticks(rotation=90)
g1.set_axis_labels("Harvest Year", "Total Cup Points") 
```

From this plot, it looks like the 2021/2022 harvest season resulted in a lot of high quality coffee. However, in our initial analysis we discovered that the 2021/2022 harvest season has the most amount of coffees in our data set -  97 to be exact which is more than twice as much as the season with the second most coffees, 2022. Moreover, there is massive spread in the 2021/2022 harvest season across coffee quality. Therefore, I believe it is fair to forgo harvest year as an indicator of quality.

```python
# Create a categorical plot in seaborn using the rating categories created above vs color

sns.set(style="ticks")
g1 = sns.catplot(x='Color', y='Total Cup Points', hue="Rating Category", data=df)
plt.xticks(rotation=90)
g1.set_axis_labels("Color", "Total Cup Points") 
```

```python
# Group colors better and fix typos

df.loc[df['Color'].str.contains('yellow green|yellow- green|yello-green'), 'Color'] = 'yellow-green'
df.loc[df['Color'].str.contains('browish-green'), 'Color'] = 'brownish-green'
df.loc[df['Color'].str.contains('bluish-green'), 'Color'] = 'blue-green'
```

```python
df['Color'].value_counts(dropna=False)
```

```python
# Create a categorical plot in seaborn using the rating categories created above vs color after updating color column

sns.set(style="ticks")
g1 = sns.catplot(x='Color', y='Total Cup Points', hue="Rating Category", data=df)
plt.xticks(rotation=90)
g1.set_axis_labels("Color", "Total Cup Points") 
```

Similar to our analysis of the harvest year plot, this plot looks to lead the same result. We could argue that green colored coffee results in a higher quality rating, however, it also has some of the lowest ratings. Therefore, we won't use color as an indicator of coffee quality.
